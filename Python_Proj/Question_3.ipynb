```python
#d.	What were the sales of the restaurant during the entire period of the dataset? 

#replaceing the dollar sign

df['item_price'] = df['item_price'].str.replace('$', '')
```

```python
df["item_price"] = pd.to_numeric(df['item_price'])
```

```python

df["Sales"] = df["quantity"] * df["item_price"]

df.head()
```
