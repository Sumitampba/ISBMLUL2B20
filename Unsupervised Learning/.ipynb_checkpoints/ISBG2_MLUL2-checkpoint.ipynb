```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

def find_fastest_improving_country(df_region):
    # Calculating here  percentage change between 2015 and 1990 for each energy type within dataframe such as (Fossil Usage,Combustible Energy,Alternate Energy)
    df_region['fu_change'] = ((df_region['fu_2015'] - df_region['fu_1990']) / df_region['fu_1990']) * 100
    df_region['comb_change'] = ((df_region['comb_2015'] - df_region['comb_1990']) / df_region['comb_1990']) * 100
    df_region['ane_change'] = ((df_region['ane_2015'] - df_region['ane_1990']) / df_region['ane_1990']) * 100

    # Create a new column to represent the overall change in energy consumption 
    df_region['overall_change'] = df_region[['fu_change', 'comb_change', 'ane_change']].mean(axis=1)

    # Find the country with the highest overall change(Metric to determine the improving state by taking mean of the Energy utilisation )
    fastest_improving_country = df_region.loc[df_region['overall_change'].idxmax()]

    # Return the result
    return fastest_improving_country['country_name']

# Example usage:
fastest_country_europe = find_fastest_improving_country(df_europe)
fastest_country_asia = find_fastest_improving_country(df_asia)
fastest_country_africa = find_fastest_improving_country(df_africa)
#fastest_country_oceania = find_fastest_improving_country(df_Africa_Oceania_north_america)


print(f"The fastest improving country in Europe is: {fastest_country_europe}")
print(f"The fastest improving country in Asia is: {fastest_country_asia}")
print(f"The fastest improving country in Africa is: {fastest_country_africa}")
#print(f"The fastest improving country in Africa_Oceania_north_america is: {fastest_country_Africa_Oceania_north_america}")


```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_for_clustering)

# Apply PCA
pca = PCA(n_components=0.95)
principal_components = pca.fit_transform(scaled_data)


# Create a DataFrame with principal components
components = pd.DataFrame(principal_components, columns=[f'PC{i}' for i in range(1, pca.n_components_ + 1)])

# Print the explained variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)

# Display the components DataFrame
print(components)
```


Principal Components (PCs): The matrix below provides the weights (loadings) of each original feature for each principal component. These loadings indicate how much each original feature contributes to each principal component. The rows represent the original features, and the columns represent the principal components.
These results are crucial for understanding how much information is captured by each principal component and how the original features contribute to each principal component. The high loadings in absolute value indicate strong contributions of the corresponding original features to the principal components.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Now data_for_clustering is our feature matrix
# Apply K-Means clustering on data_for_clustering
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_labels = kmeans.fit_predict(data_for_clustering)

# define pipeline steps
pca_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dim_reduction', PCA(n_components=2))
])

# Fit and transform data using the pipeline
pca_results = pca_pipeline.fit_transform(data_for_clustering)

# create the scatter plot
fig, ax = plt.subplots(figsize=(16, 11))
scatter = ax.scatter(
    x=pca_results[:, 0],
    y=pca_results[:, 1],
    c=cluster_labels,  # Use cluster labels for coloring
    cmap=plt.cm.get_cmap('Spectral'),
    alpha=0.4
)

# produce a legend with the colors from the scatter
legend = ax.legend(*scatter.legend_elements(), title="Clusters", bbox_to_anchor=(1.05, 1), loc='upper left')
ax.add_artist(legend)
ax.set_title("PCA visualization with Clusters")
plt.xlabel("1st principal component")
plt.ylabel("2nd principal component")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering



# Find the best clustering algorithm and number of clusters using silhouette score
for algorithm in ['KMeans', 'Agglomerative']:
    if algorithm == 'KMeans':
        # KMeans clustering
        for k in range(2, 6):
            kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
            df['Cluster'] = kmeans.fit_predict(data_for_clustering)
            silhouette_avg = silhouette_score(data_for_clustering, df['Cluster'])
            print(f"Silhouette Score for {k} clusters ({algorithm}): {silhouette_avg}")

    elif algorithm == 'Agglomerative':
        # Agglomerative clustering
        for k in range(2, 6):
            agglomerative = AgglomerativeClustering(n_clusters=k)
            df['Cluster'] = agglomerative.fit_predict(data_for_clustering)
            silhouette_avg = silhouette_score(data_for_clustering, df['Cluster'])
            print(f"Silhouette Score for {k} clusters ({algorithm}): {silhouette_avg}")
```

t-distributed Stochastic Neighbor Embedding (t-SNE) is a dimensionality reduction technique commonly used for visualizing high-dimensional data in lower-dimensional space. It is particularly effective in preserving the local structure and relationships between data points. 

```python
from sklearn.manifold import TSNE
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA



# Define the pipeline
tsne_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dim_reduction', PCA(n_components=4)),
    ('2d_reduction', TSNE(n_components=2, init='pca', random_state=42))
])

# Transform the data using the pipeline
tsne_results = tsne_pipeline.fit_transform(data_for_clustering)

# Create the scatter plot
fig, ax = plt.subplots(figsize=(16, 11))
scatter = ax.scatter(
    x=tsne_results[:, 0],
    y=tsne_results[:, 1],
    c=df['Cluster'],  # Assuming 'Cluster' is the cluster labels column
    cmap=plt.cm.get_cmap('Spectral'),
    alpha=0.4
)

# Produce a legend with the colors from the scatter
legend = ax.legend(*scatter.legend_elements(), title="Classes", bbox_to_anchor=(1.05, 1), loc='upper left')
ax.add_artist(legend)
ax.set_title("t-SNE visualization")
plt.xlabel("1st t-SNE component")
plt.ylabel("2nd t-SNE component")
plt.show()
```

```python
from sklearn.manifold import TSNE
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Define the pipeline
tsne_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dim_reduction', PCA(n_components=4)),
    ('2d_reduction', TSNE(n_components=2, init='pca', random_state=50))
])

# Transform the data using the pipeline
tsne_results = tsne_pipeline.fit_transform(data_for_clustering)

# Create the scatter plot
fig, ax = plt.subplots(figsize=(16, 11))
scatter = ax.scatter(
    x=tsne_results[:, 0],
    y=tsne_results[:, 1],
    c=df['Cluster'],  # Assuming 'Cluster' is the cluster labels column
    cmap=plt.cm.get_cmap('Spectral'),
    alpha=0.4
)

# Annotate each point with its corresponding country name
for i, country_name in enumerate(df['country_name'].tolist()):
    ax.annotate(country_name, (tsne_results[i, 0], tsne_results[i, 1]), textcoords='offset points')


# Produce a legend with the colors from the scatter
legend = ax.legend(*scatter.legend_elements(), title="Classes", bbox_to_anchor=(0.75, 1), loc='upper left')
ax.add_artist(legend)
ax.set_title("t-SNE visualization with Country Names")
plt.xlabel("1st t-SNE component")
plt.ylabel("2nd t-SNE component")
plt.show()
```

Anomaly Detection using Isolation Forest 

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

features = ['total_ep_1990', 'total_ep_2015', 'total_co2_emi_2015', 'total_co2_emi_2020', 'carbon_intensity_1990', 'carbon_intensity_2014']
X = df[features]

# Normalize/Scale data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
anomaly_labels = isolation_forest.fit_predict(X_scaled)

df['anomaly_label'] = anomaly_labels

plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(x='total_ep_2015', y='total_co2_emi_2015', hue='anomaly_label', data=df, palette={1: 'blue', -1: 'red'}, s=100)
anomalies = df[df['anomaly_label'] == -1]
for i, country_name in anomalies['country_name'].items():
    scatter.annotate(country_name, (df.loc[i, 'total_ep_2015'], df.loc[i, 'total_co2_emi_2015']), textcoords='offset points', xytext=(0, 5), ha='left')

plt.title('Isolation Forest - Anomaly Detection')
plt.xlabel('Total Energy Production 2015')
plt.ylabel('Total Co2 Emission 2015')
plt.legend(title='Anomaly Label', loc='upper right')
plt.show()
print("Anomalies:")
print(anomalies)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity

# Select relevant columns
columns_for_comparison = ['total_ep_2015', 'total_eu_2015', 'eu_per_capita_2015', 'total_co2_emi_2015', 'co2_emi_per_capita_2015']

# Extract features for comparison
features_for_comparison = df[columns_for_comparison]

# Check if the DataFrame is not empty
if not features_for_comparison.empty:
    # Normalize the data (optional for Euclidean distance)
    scaler = StandardScaler()
    features_normalized = scaler.fit_transform(features_for_comparison)

    # Compute pairwise Euclidean distances
    euclidean_distances_matrix = euclidean_distances(features_normalized)

    # Compute pairwise cosine similarities
    cosine_similarity_matrix = cosine_similarity(features_for_comparison)

    # Choose a pair of countries for demonstration
    country1 = 'India'
    country2 = 'Australia'

    # Check if the specified countries exist in the DataFrame
    if country1 in df['country_name'].values and country2 in df['country_name'].values:
        # Find the indices of the selected countries
        index_country1 = df[df['country_name'] == country1].index[0]
        index_country2 = df[df['country_name'] == country2].index[0]

        # Retrieve the Euclidean distance and cosine similarity between the selected countries
        euclidean_distance = euclidean_distances_matrix[index_country1, index_country2]
        cosine_similarity_value = cosine_similarity_matrix[index_country1, index_country2]

        # Print the results
        print(f"Euclidean Distance between {country1} and {country2}: {euclidean_distance}")
        print(f"Cosine Similarity between {country1} and {country2}: {cosine_similarity_value}")
    else:
        print("One or both of the specified countries not found in the DataFrame.")
else:
    print("DataFrame is empty.")
```

The Euclidean distance and cosine similarity are two different measures used to quantify the similarity or dissimilarity between two vectors (in this case, feature vectors representing different countries). Here's a detailed report on the interpretation of the results:

Euclidean Distance:

The Euclidean distance is a measure of the straight-line distance between two points in space.
In the context of feature vectors, a lower Euclidean distance indicates greater similarity between the vectors.
In your case, the Euclidean Distance between Australia and India is approximately 0.5274.
Interpretation: The relatively small Euclidean distance suggests that Australia and India have similar feature profiles based on the selected features ('total_ep_2015', 'total_eu_2015', 'eu_per_capita_2015', 'total_co2_emi_2015', 'co2_emi_per_capita_2015').
Cosine Similarity:

Cosine similarity measures the cosine of the angle between two non-zero vectors.
It ranges from -1 (completely dissimilar) to 1 (completely similar), with 0 indicating orthogonality (no similarity).
In your case, the Cosine Similarity between Australia and India is approximately 0.9966.
Interpretation: The high cosine similarity value close to 1 indicates a strong similarity between the feature vectors of Australia and India. The vectors are nearly parallel in the multi-dimensional space, suggesting that these countries have similar trends in the selected features.
Overall Interpretation:

Both the Euclidean distance and cosine similarity metrics suggest that Australia and India have similar patterns in the selected energy-related features. The small Euclidean distance and high cosine similarity value reinforce the idea that these countries exhibit comparable characteristics in terms of energy production, EU-related features, and CO2 emission statistics.
