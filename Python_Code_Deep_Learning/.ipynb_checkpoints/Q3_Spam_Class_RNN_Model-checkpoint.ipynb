```python
import pandas as pd
import re
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import seaborn as sns
```

```python
def remove_punctuation_regex(text):
    return re.sub(r'[^\w\s]', '', text)
```

```python
def char_count(str, chrs):
    count = 0
    for i in str: 
        for chr in chrs:
            if i == chr:
                count += 1
    return count
```

```python
def count_upper_words(str):
    count = 0
    for i in str.split(" "):
        if i.isupper():
            count+=1
    return count
```

```python
# s.isdigit()
def count_digits_words(str):
    count = 0
    for i in str.split(" "):
        if i.isdigit():
            count+=1
    return count
```

```python
def find_links(str):
    count = 0
    for i in str.split(" "): 
        if i[:4] in ["www.","http"]:
            count += 1
    return count
```

```python
def add_flagged_values(str):
    score = 0
    for i in str.split(" "):
        if i in sorted_freq.keys():
            score += sorted_freq[i]
    return score
```

```python
df["length_of_message"] = [len(i) for i in df["Message"]]
df["exclaimation"] = [char_count(i, ["!"]) for i in df["Message"]]
df["currency"] = [char_count(i, ["$","£","€"]) for i in df["Message"]] 
df["all_upper"] = [count_upper_words(i) for i in df["Message"]] 
df["all_digit"] = [count_digits_words(i) for i in df["Message"]] 
df["links"] = [find_links(i) for i in df["Message"]] 
df["length_of_message"]
```

```python
df["Message"] = [remove_punctuation_regex(i).lower() for i in df["Message"]]
df["Message"] = [re.sub(r'\d+', '', i) for i in df["Message"]]

df["Message"]
```

```python
all_spam = ""
for i in df[df["Category"]=="spam"]["Message"]:
    all_spam += i

all_spam
```

```python
frequency = {}
words = all_spam.lower().split(" ")
for word in words:
    word = word.strip().lower()
    frequency[word] = frequency.get(word, 0) + 1
words
```

```python
en_stops = set(stopwords.words('english'))

# Here we sor the dictionary based on how frequently they appear
sorted_freq = dict(sorted(frequency.items(), key=lambda item: item[1], reverse=True))
sorted_freq
```

```python
# This block of code does 3 things:
#     1. takes out stop words
#     2. takes out words that 1 or 2 letters long
#     3. takes out words that appear less than 10 times
    
for i in list(sorted_freq.keys()):
    if i in en_stops:
        del sorted_freq[i]
    elif len(i)<3:
        del sorted_freq[i]
    elif sorted_freq[i] < 10:
        del sorted_freq[i]
```

```python
# here is a vew of the dictionary created above in the form of a dataframe
pd.DataFrame.from_dict(sorted_freq, orient='index').reset_index().rename(columns={"index":"word",0:"frequency"}).head(15)
```

```python
# Below is a wordcloud, a good visual representation of how often each word appears in the spam text#
wc = WordCloud().generate_from_frequencies(sorted_freq)
plt.imshow(wc)
plt.axis("off")
```

```python
# Here I create a new field based on the word frequencies generated above:
#     1. each word in the dictionary above is assigned points based on how frequently it appears in the spam text
#     2. we look at each message and sum all the points for each word that appears in that message
#     3. we normalize the total score by 1000

df["flagged_words_points"] = [add_flagged_values(i) for i in df["Message"]] 
df["flagged_words_points"] = df["flagged_words_points"]/1000
df["flagged_words_points"]
```

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Load the dataset from CSV file
file_path = r'C:\Users\sumit\Desktop\ISB\Classwork\Deep Learning_R4\Python_Code\SPAM text message 20170820 - Data (1).csv'
df = pd.read_csv(file_path, encoding='latin-1')

# Select relevant columns and rename them
df = df[['Category', 'Message']]
df.columns = ['label', 'message']

# Convert labels to binary (0 for ham, 1 for spam)
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Tokenize the messages
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['message'])
sequences = tokenizer.texts_to_sequences(df['message'])

# Pad sequences to a fixed length
max_len = 100  # Choose the maximum sequence length
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post', truncating='post')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, df['label'], test_size=0.2, random_state=30)

# Check vocabulary size (number of unique words)
vocab_size = len(tokenizer.word_index) + 1
print(f"Vocabulary size: {vocab_size}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = r'C:\Users\sumit\Desktop\ISB\Classwork\Deep Learning_R4\Python_Code\SPAM text message 20170820 - Data (1).csv'  # Update with the path to your CSV file containing the dataset
df = pd.read_csv(file_path, encoding='latin-1')

# Step 4: Build the RNN model
embedding_dim = 50  # You can adjust the embedding dimension based on your vocabulary size
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim))
model.add(SimpleRNN(64))  # No need to specify input_length here
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Step 5: Train the model
num_epochs = 15
batch_size = 32
history = model.fit(X_train, y_train, epochs=num_epochs, batch_size=batch_size, validation_split=0.2)

# Step 6: Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")

# Step 7: Plot training history (optional)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
