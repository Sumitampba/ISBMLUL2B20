```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Replace 'your_dataset.csv' with the actual file path of your downloaded dataset
dataset_path = r'C:\Users\sumit\Desktop\ISB\Classwork\MLSL_R3\tmnst DATA SET.csv'
df = pd.read_csv(dataset_path)

# Normalize pixel values to [0, 1]
df.iloc[:, 1:] = df.iloc[:, 1:] / 255.0

# Function to perform PCA vs Fisher scatter plots for 3D visualization
def pca_vs_fisher_3d(train, classes):
    # Extract data for the specified classes
    subset_data = train[train['labels'].isin(classes)]

    # Separate features and labels
    X = subset_data.iloc[:, 1:].values
    y = subset_data['labels'].values

    # Convert class labels to numeric values
    label_encoder = LabelEncoder()
    y_numeric = label_encoder.fit_transform(y)

    # Part 1: PCA PROJECTION
    pca = PCA(n_components=3)
    pca_result = pca.fit_transform(X)

    # Plot PCA projection
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(121, projection='3d')
    scatter = ax.scatter(pca_result[:, 0], pca_result[:, 1], pca_result[:, 2], c=y_numeric, cmap='viridis')
    ax.set_title('PCA Projection')
    ax.set_xlabel('Principal Component 1')
    ax.set_ylabel('Principal Component 2')
    ax.set_zlabel('Principal Component 3')
    plt.legend(*scatter.legend_elements(), title='Classes')
    plt.show()

    # Part 2: FISHER PROJECTION
    lda = LinearDiscriminantAnalysis(n_components=3)
    fisher_result = lda.fit_transform(X, y_numeric)

    # Plot Fisher projection
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(122, projection='3d')
    scatter = ax.scatter(fisher_result[:, 0], fisher_result[:, 1], fisher_result[:, 2], c=y_numeric, cmap='viridis')
    ax.set_title('Fisher Projection')
    ax.set_xlabel('Fisher Component 1')
    ax.set_ylabel('Fisher Component 2')
    ax.set_zlabel('Fisher Component 3')
    plt.legend(*scatter.legend_elements(), title='Classes')
    plt.show()

# Split data into train and test (70% train, 30% test)
train, test = train_test_split(df, test_size=0.3, stratify=df['labels'], random_state=42)

# Call the function for three different combinations of classes
classes_combination_1 = ['D', 'F', 'J', 'H']
classes_combination_2 = ['A', 'B', 'C', 'E']
classes_combination_3 = ['G', 'K', 'M', 'O']

pca_vs_fisher_3d(train, classes_combination_1)
pca_vs_fisher_3d(train, classes_combination_2)
pca_vs_fisher_3d(train, classes_combination_3)
```
